## 3- Feature Engineering:

```python
# Create Total Spend per Customer, total amount spent on each transaction
df['TotalSpend'] = df['Quantity'] * df['Price']
```

```python
# Group by 'CustomerID' and aggregate multiple metrics
customer_group = df.groupby('Customer ID').agg({
    'TotalSpend': 'sum', # Sum of total spend per customer
    'Invoice': 'nunique',  # Number of unique orders per customer
    'Quantity': 'sum', # Total quantity of items purchased by customer
    'InvoiceDate': ['max', 'min'] # Most recent and earliest purchase dates
}).reset_index()
```

```python
# Rename columns for better readability
customer_group.columns = ['Customer ID', 'TotalSpend', 'NumOrders', 'TotalQuantity', 'LastPurchase', 'FirstPurchase']
```

```python
# Average Order Value, representing the average amount spent per order by each customer.
customer_group['AvgOrderValue'] = customer_group['TotalSpend'] / customer_group['NumOrders']
```

```python
# Number of unique products purchased
unique_products = df.groupby('Customer ID')['StockCode'].nunique().reset_index()
unique_products.columns = ['Customer ID', 'UniqueProducts']

# Merge the unique products data with the customer group data
customer_group = pd.merge(customer_group, unique_products, on='Customer ID', how='left')
```

```python
# Recency (days since last purchase)
##Calculate the maximum date in the dataset
max_date = df['InvoiceDate'].max()

# Calculate the recency for each customer (days since last purchase)
#max_date is the most recent date in the dataset, and recency is the difference between this date and the customer's last purchase date (LastPurchase)
customer_group['Recency'] = (max_date - customer_group['LastPurchase']).dt.days
```

```python
# Calculate the frequency (average number of days between orders)
customer_group['Frequency'] = (customer_group['LastPurchase'] - customer_group['FirstPurchase']).dt.days / customer_group['NumOrders']
```

```python
# Customer Lifetime Value (using Total Spend as a simple proxy)
customer_group['CLV'] = customer_group['TotalSpend']
```

```python
# Function to detect outliers
def detect_outliers(df, n, features_list):
    outlier_indices = [] 
    for feature in features_list: 
        # Calculate Q1, Q3, and IQR
        Q1 = np.percentile(df[feature], 25)
        Q3 = np.percentile(df[feature], 75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR 
        
        # Identify Outliers
        outlier_list_col = df[(df[feature] < Q1 - outlier_step) | (df[feature] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col) 
    
    # Count the Occurrences of Each Outlier
    outlier_indices = Counter(outlier_indices)
    
    # Filter Out Multiple Outliers
    multiple_outliers = list(key for key, value in outlier_indices.items() if value > n) 
    return multiple_outliers
```

```python
# Detect outliers
outliers_to_remove = detect_outliers(customer_group, 2, numerical_features)
```

```python
print("We will drop these {} indices: ".format(len(outliers_to_remove)), outliers_to_remove)
```

```python
# Remove detected outliers from the DataFrame
customer_group_cleaned = customer_group.drop(outliers_to_remove, axis=0).reset_index(drop=True)
```

# 5- Data Preparation for Clustering:

```python
# Features for clustering
features = ['TotalSpend', 'NumOrders', 'AvgOrderValue', 'UniqueProducts', 'Recency', 'Frequency']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_group_cleaned[features])
```

# 6- Dimensionality Reduction

```python
# Apply PCA, ,it reduces the dimensionality of the dataset while retaining as much variance as possible.
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
```

```python
# Visualize the data in 2D
plt.figure(figsize=(10, 7))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Customer Data')
plt.show()
```

# 7- Clustering:

- a) K-means

```python
# Determine the optimal number of clusters using the Elbow method, 
#Run cluster analysis as 1 till 12, we store WCSS stores. This WSS will help us in elbow method to find optimal number of clusters
```

```python
K = range(1,12)
wcss = []
for i in K:
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
```

```python
#store number of clusters along with their wcss
mycenters = pd.DataFrame({'Clusters ' : K, 'WCSS' : wcss})
mycenters
```

```python
#plot elbow method 
plt.plot(range(1, 12), wcss, marker = 'o')
plt.title('The Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

- Here we can notice that the optimal number of clustering is 7,the point after which the distortion/inertia starts decreasing in a linear fashion.

```python
#check by Silhouette Score Analysis
for i in range(2,11):
    labels = KMeans(n_clusters = i, init = 'k-means++', random_state = 42).fit(scaled_features).labels_
    print("Silhouette Score fo k(clusters) =" +str(i) +" is " +str(metrics.silhouette_score(scaled_features,labels,metric="euclidean", sample_size =1000,random_state = 42)))
```

```python
# Ensure you use the same data for both clustering and assigning labels
kmeans = KMeans(n_clusters=7, init='k-means++', random_state=42)
customer_group_cleaned['KMeansCluster'] = kmeans.fit_predict(scaled_features)
```

```python
### Calculating the performance of the clusters

silhouette_score_kmeans = round(silhouette_score(scaled_features, customer_group_cleaned['KMeansCluster']),3)
calinski_harabasz_score_kmeans = round(calinski_harabasz_score(scaled_features, customer_group_cleaned['KMeansCluster']),3)

print('Silhouette Score : {}'.format(silhouette_score_kmeans))
print('Calinski Harabasz Score : {}'.format(calinski_harabasz_score_kmeans))
```

- b) Hierarchical Clustering:

```python
### Using the dendrogram to find the optimal number of clusters
dendrogram = sch.dendrogram(sch.linkage(scaled_features, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

We can notice here that we have 3 clusters as optimal number

```python
agg_clustering = AgglomerativeClustering(n_clusters=3)
customer_group_cleaned['HierarchicalCluster'] = agg_clustering.fit_predict(scaled_features)
```

```python
### Calculating the performance of the clusters

silhouette_score_hierarchical = round(silhouette_score(scaled_features, customer_group_cleaned['HierarchicalCluster']),3)
calinski_harabasz_score_hierarchical = round(calinski_harabasz_score(scaled_features, customer_group_cleaned['HierarchicalCluster']),3)

print('Silhouette Score : {}'.format(silhouette_score_hierarchical))
print('Calinski Harabasz Score : {}'.format(calinski_harabasz_score_hierarchical))
```

- c) DBSCAN Clustering:

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm that can identify clusters of varying shapes and sizes, and it also labels some points as noise. However, the Calinski-Harabasz Index (also known as the Variance Ratio Criterion) is generally not suitable for evaluating DBSCAN results

```python
from sklearn.cluster import DBSCAN
# Fit DBSCAN (choose epsilon and min_samples based on experimentation)
dbscan = DBSCAN(eps=0.5, min_samples=10)
customer_group_cleaned['DBSCANCluster'] = dbscan.fit_predict(scaled_features)
```

```python
#Points within a radius of 0.5 units are considered neighbors.
#A point needs to have at least 10 neighbors (including itself) to be considered a core point.
```

```python
### Calculating the performance of the clusters

silhouette_score_DBSCAN = round(silhouette_score(scaled_features, customer_group_cleaned['DBSCANCluster']),3)
print('Silhouette Score : {}'.format(silhouette_score_DBSCAN))
```
